```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# date range from October 1, 2020 until yesterday
dates = pd.date_range(start='10/1/2020', end=datetime.now() - timedelta(1))

unix_dates = [int(date.timestamp()) for date in dates]


# define a function to get data between two unix timestamps
def get_data(from_date, to_date):
    url = f'https://eodhd.com/api/intraday/NSEBANK.INDX?api_token=65e185a7d8ff47.50324444&fmt=json&from={from_date}&to={to_date}'
    return requests.get(url).json()


# initialize an empty DataFrame
df = pd.DataFrame()

# handle the 600 days limit
for i in range(0, len(unix_dates), 600):
    from_date = unix_dates[i]
    to_date = unix_dates[i + 599] if i + 599 < len(unix_dates) else unix_dates[-1]
    data = get_data(from_date, to_date)

    # Convert data to DataFrame and append to main DataFrame
    temp_df = pd.DataFrame(data)
    df = pd.concat([df, temp_df], ignore_index=True)

print(df)
```

```python
import pandas as pd
from pytz import timezone

# Assuming df is your DataFrame
# Drop the first two columns
df = df.drop(df.columns[[0, 1]], axis=1)

# Convert 'datetime' from GMT to IST
# Ensure 'datetime' is a datetime object
df['datetime'] = pd.to_datetime(df['datetime'])
# Set timezone to GMT
df['datetime'] = df['datetime'].dt.tz_localize('GMT')
# Convert to IST
df['datetime'] = df['datetime'].dt.tz_convert('Asia/Kolkata')

# Set 'datetime' as the index
df = df.set_index('datetime')

```
